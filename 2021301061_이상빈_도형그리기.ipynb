Q1. 마우스 버튼 클릭 대신에 드래그로 도형을 그리려면?

```python
import numpy as np
import cv2

def onMouse(event, x, y, flags, param):
    global title, pt

    if event == cv2.EVENT_LBUTTONDOWN:
        if pt[0] < 0:
            pt = (x, y)
    
    elif event == cv2.EVENT_RBUTTONDOWN:
        if pt[0] < 0:
            pt = (x, y)
            
    elif event == cv2.EVENT_LBUTTONUP:
        if pt[0] >= 0:
            cv2.rectangle(image, pt, (x, y), (255,0,0), 2)
            cv2.imshow(title, image)
            pt = (-1, -1)
            
    elif event == cv2.EVENT_RBUTTONUP:
        if pt[0] >= 0:
            dx, dy = (pt[0] - x)/2, (pt[1] - y)/2
            radius = int(np.sqrt(dx * dx + dy * dy))
            center=int(pt[0] - dx), int(pt[1] - dy)
            cv2.circle(image, center, radius, (0,0,255), 2)
            cv2.imshow(title, image)
            pt = (-1, -1)

image = np.full((300, 500, 3), (255, 255, 255), np.uint8)

pt = (-1, -1)
title = "Draw & Drag Event"
cv2.imshow(title, image)
cv2.setMouseCallback(title, onMouse)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

Q2. 마우스를 드래그할 때 러버밴드(rubber band) 효과를 넣어서 그리려면?

```python
import numpy as np
import cv2

def onMouse(event, x, y, flags, param):
    global title, pt, line_pt, line_image

    if event == cv2.EVENT_LBUTTONDOWN:
        if line_pt[0] < 0:
            pt = (x, y)
            line_pt = (x, y)
            line_image = image.copy()

    elif event == cv2.EVENT_RBUTTONDOWN:
        if line_pt[0] < 0:
            pt = (x, y)
            line_pt = (x, y)
            line_image = image.copy()

    elif event == cv2.EVENT_MOUSEMOVE:
        if flags & cv2.EVENT_FLAG_LBUTTON:
            cv2.line(line_image, line_pt, (x, y), (0,0,0), 2)
            cv2.imshow(title, line_image)
            line_pt = (x, y)
        elif flags & cv2.EVENT_FLAG_RBUTTON:
            cv2.line(line_image, line_pt, (x, y), (0,0,0), 2)
            cv2.imshow(title, line_image)
            line_pt = (x, y)
            
    elif event == cv2.EVENT_LBUTTONUP:
        if pt[0] >= 0:
            cv2.rectangle(image, pt, (x, y), (255,0,0), 2)
            cv2.imshow(title, image)
            line_pt = (-1, -1)
            
    elif event == cv2.EVENT_RBUTTONUP:
        if pt[0] >= 0:
            dx, dy = (pt[0] - x)/2, (pt[1] - y)/2
            radius = int(np.sqrt(dx * dx + dy * dy))
            center = int(pt[0] - dx), int(pt[1] - dy)
            cv2.circle(image, center, radius, (0,0,255), 2)
            cv2.imshow(title, image)
            line_pt = (-1, -1)

image = np.full((300, 500, 3), (255, 255, 255), np.uint8)

pt = (-1, -1)
line_pt = (-1, -1)
title = "RubberBand Event"
cv2.imshow(title, image)
cv2.setMouseCallback(title, onMouse)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
