### ex4-5 오류 관련
- add_value = value - int(image[0][0])에서 int를 뺌<br>
  => add_value = value - image[0][0]
- 하지만 value는 dtype이 int이고 image[0][0]의 dtype은 uint8이라 warnings 발생(예제의 오류보다는 나음)
- 저자는 이 오류를 해결하기 위해서 image[0][0]에 int를 씌워서 해결하려 함
- image = image + add_value if add_value > 0 else image - (- add_value) (<- img = img + add_value)

```python
import numpy as np
import cv2

def onChange(value):
    global img, title
    
    add_value = value - img[0][0]
    print("추가 화소값: ", add_value)

    # image += add_value
    img = img + add_value if add_value > 0 else img - (- add_value)
    cv2.imshow(title, img)

img = np.zeros((300, 500), np.uint8)

title = 'Trackbar Event'
cv2.imshow(title, img)

cv2.createTrackbar('Brightness', title, img[0][0], 255, onChange)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import numpy as np
import cv2

def onChange(value):
    global image, title                                 # 전역 변수 참조

    add_value = value - image[0][0]                # 트렉바 값과 영상화소값 차분
    image = image + add_value if add_value > 0 else image - (- add_value)
    cv2.imshow(title, image)

def onMouse(event, x, y, flags, param):                 # 마우스 콜백 함수
    global image, bar_name
    
    if event == cv2.EVENT_RBUTTONDOWN:
        if (image[0][0] < 246): image += 10
        cv2.setTrackbarPos(bar_name, title, image[0][0])		# 트랙바 위치 변경 
        cv2.imshow(title, image)
        
    elif event == cv2.EVENT_LBUTTONDOWN :
        if (image[0][0] >= 10): image -= 10
        cv2.setTrackbarPos(bar_name, title, image[0][0])		# 트랙바 위치 변경 
        cv2.imshow(title, image)

image = np.zeros((300, 500), np.uint8)                 

title = "Trackbar & Mouse Event"                    		# 윈도우 이름
bar_name = "Brightness"                                # 트랙바 이름
cv2.imshow(title, image)    

cv2.createTrackbar(bar_name, title, image[0][0], 255, onChange)   # 트랙바 콜백 함수
cv2.setMouseCallback(title, onMouse)
cv2.waitKey(0)													# 키 입력 대기
cv2.destroyAllWindows()                                	# 모든 윈도우 닫기
```

```python
import numpy as np
import cv2

def onChange(value):
    global image, title
    add_value = value - int(image[0][0])
    image = image + add_value if add_value > 0 else image - (-add_value)
    cv2.imshow(title, image)

def onMouse(event, x, y, flags, param):
    global image, bar_name, title

    if event == cv2.EVENT_RBUTTONDOWN:
        if image[0][0] < 246:
            image += 10
            cv2.setTrackbarPos(bar_name, title, image[0][0])
            cv2.imshow(title, image)

    elif event == cv2.EVENT_LBUTTONDOWN:
        if image[0][0] >= 10:
            image -= 10
            cv2.setTrackbarPos(bar_name, title, image[0][0])
            cv2.imshow(title, image)

image = np.zeros((300, 500), np.uint8)
title = "Trackbar & Mouse & Keyboard Event"
bar_name = "Brightness"

cv2.imshow(title, image)
cv2.createTrackbar(bar_name, title, image[0][0], 255, onChange)
cv2.setMouseCallback(title, onMouse)

while True:
    key = cv2.waitKeyEx(100)

    if key == 27:  # ESC 종료
        break
    elif key == 0x250000:  # ← 왼쪽 키
        if image[0][0] >= 10:
            image -= 10
            cv2.setTrackbarPos(bar_name, title, image[0][0])
            cv2.imshow(title, image)
    elif key == 0x270000:  # → 오른쪽 키
        if image[0][0] < 246:
            image += 10
            cv2.setTrackbarPos(bar_name, title, image[0][0])
            cv2.imshow(title, image)

cv2.destroyAllWindows()
```

```python
# ex. 4-11
import numpy as np
import cv2

def onMouse(event, x, y, flags, param):
    global title, pt                            # 전역 변수 참조

    if event == cv2.EVENT_LBUTTONDOWN:
        if pt[0] < 0:
            pt = (x, y)                         # 시작 좌표 지정
        else:
            cv2.rectangle(image, pt, (x, y), (255,0,0), 2)
            cv2.imshow(title, image)
            pt = (-1, -1)                       # 시작 좌표 초기화
            
    elif event == cv2.EVENT_RBUTTONDOWN:
        if pt[0] < 0: pt = (x, y)
        else:
            dx, dy = pt[0] - x, pt[1] - y       # 두 좌표 간의 거리
            radius = int(np.sqrt(dx * dx + dy * dy))
            cv2.circle(image, pt, radius, (0,0,255), 2)
            cv2.imshow(title, image)
            pt = (-1, -1)                       # 시작 좌표 초기화

image = np.full((300, 500, 3), (255, 255, 255), np.uint8) # 흰색 배경 영상

pt = (-1, -1)                                   # 시작 좌표 초기화
title = "Draw Event"
cv2.imshow(title, image)                        # 윈도우에 영상 띄우기
cv2.setMouseCallback(title, onMouse)            # 마우스 콜백 함수 등록
cv2.waitKey(0)
cv2.destroyAllWindows()
```

Q1. 마우스 버튼 클릭 대신에 드래그로 도형을 그리려면?
- LBUTTONUP, RBUTTONUP 사용

```python
import numpy as np
import cv2

def onMouse(event, x, y, flags, param):
    global title, pt

    if event == cv2.EVENT_LBUTTONDOWN:
        if pt[0] < 0:
            pt = (x, y)
            
    elif event == cv2.EVENT_RBUTTONDOWN:
        if pt[0] < 0:
            pt = (x, y)
            
    elif event == cv2.EVENT_LBUTTONUP:
        cv2.rectangle(image, pt, (x, y), (255,0,0), 2)
        cv2.imshow(title, image)
        pt = (-1, -1)

    elif event == cv2.EVENT_RBUTTONUP:
        dx, dy = (pt[0] - x)/2, (pt[1] - y)/2
        radius = int(np.sqrt(dx * dx + dy * dy))
        center=int(pt[0] - dx), int(pt[1] - dy)
        cv2.circle(image, center, radius, (0,0,255), 2)
        cv2.imshow(title, image)
        pt = (-1, -1)

image = np.full((300, 500, 3), (255, 255, 255), np.uint8)

pt = (-1, -1)
title = "Draw Event"
cv2.imshow(title, image)
cv2.setMouseCallback(title, onMouse)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

Q2. 마우스를 드래그할 때 러버밴드(rubber band) 효과를 넣어서 그리려면?
- 마우스 move 이벤트
- 드래그할 때는 검은 선이 그어지고 드래그가 끝나면 그 위치에 흰색 선으로 덮음

```python
import numpy as np
import cv2

def onMouse(event, x, y, flags, param):
    global title, pt, start_pt, image, backup_image, line_image
    
    if event == cv2.EVENT_LBUTTONDOWN:
        if pt[0] < 0:
            pt = (x, y)
            start_pt = (x, y)  # 시작점을 저장하는 변수
            backup_image = image.copy() # 전체 이미지 저장
            line_image = image.copy() # 러버밴드를 그리는 이미지
            
    elif event == cv2.EVENT_RBUTTONDOWN:
        if pt[0] < 0:
            pt = (x, y)
            start_pt = (x, y)
            backup_image = image.copy()
            line_image = image.copy()
            
    elif event == cv2.EVENT_MOUSEMOVE:
        if flags & cv2.EVENT_FLAG_LBUTTON:
            cv2.line(line_image, pt, (x, y), (255, 255, 0), 2)
            cv2.imshow(title, line_image)
            pt = (x, y)
        elif flags & cv2.EVENT_FLAG_RBUTTON:
            cv2.line(line_image, pt, (x, y), (0, 255, 255), 2)
            cv2.imshow(title, line_image)
            pt = (x, y)
            
    elif event == cv2.EVENT_LBUTTONUP:
        # cv2.line(image, start_pt, (x, y), (255,255,255), 2)
        # image = backup_image.copy()  # 경로 선 제거
        cv2.rectangle(image, start_pt, (x, y), (255,0,0), 2)
        cv2.imshow(title, image)
        pt = (-1, -1)
    
    elif event == cv2.EVENT_RBUTTONUP:
        cv2.line(image, start_pt, (x, y), (255,255,255), 2)
        # image = backup_image.copy()
        dx, dy = (start_pt[0] - x)/2, (start_pt[1] - y)/2
        radius = int(np.sqrt(dx * dx + dy * dy))
        center = int(start_pt[0] - dx), int(start_pt[1] - dy)
        cv2.circle(image, center, radius, (0,0,255), 2)
        cv2.imshow(title, image)
        pt = (-1, -1)

image = np.full((300, 500, 3), (255, 255, 255), np.uint8)

pt = (-1, -1)
start_pt = (-1, -1)
title = "Draw Event"
cv2.imshow(title, image)
cv2.setMouseCallback(title, onMouse)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import numpy as np
import cv2

def onMouse(event, x, y, flags, param):
    global image, start_pt, drawing, title

    temp = image.copy()  # 매번 원본 복사

    if event == cv2.EVENT_LBUTTONDOWN or event == cv2.EVENT_RBUTTONDOWN:
        drawing = True
        start_pt = (x, y)

    elif event == cv2.EVENT_MOUSEMOVE:
        if drawing:
            cv2.line(temp, start_pt, (x, y), (0, 0, 0), 2)
            cv2.imshow(title, temp)

    elif event == cv2.EVENT_LBUTTONUP:
        drawing = False
        if pt == (x, y): return
        else:
            cv2.rectangle(image, start_pt, (x, y), (255, 0, 0), 2)
            cv2.imshow(title, image)

    elif event == cv2.EVENT_RBUTTONUP:
        drawing = False
        if pt == (x, y): return
        else:
            dx, dy = start_pt[0] - x, start_pt[1] - y
            radius = int(np.sqrt(dx**2 + dy**2))
            cv2.circle(image, start_pt, radius, (0, 0, 255), 2)
            cv2.imshow(title, image)

image = np.full((400, 600, 3), 255, np.uint8)
start_pt = (-1, -1)
drawing = False
title = "Pointer RubberBand"

cv2.imshow(title, image)
cv2.setMouseCallback(title, onMouse)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
