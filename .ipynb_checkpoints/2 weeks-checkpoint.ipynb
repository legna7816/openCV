### 크기, 초기값 생성
- np.empty(shape, [, dtype]) # 초기화 X, 초기화하려면 fill()을 사용해야 함
- np.zeros(shape, [, dtype]) # 0으로 초기화
- np.ones(shape, [, dtype])  # 1로 초기화
- np.full(shape, full_value, [, dtype])  # full_value로 초기화

```python
a=np.empty((2,3))
a
```

```python
a.dtype
```

```python
a.fill(255) # 주어진 값인 255로 채움
a
```

```python
b=np.zeros((2,3))
b
```

```python
c=np.zeros((2,3),dtype=np.uint8)
c
```

```python
d=np.ones((2,3),dtype=np.uint16)
d
```

```python
e=np.full((2,4,3),255,dtype=np.uint8)
e
```

### 기존 배열과 같은 크기 배열 생성
- np.empty_like(array, [, dtype])
- np.zeros_like(array, [, dtype])
- np.ones_like(array, [, dtype])
- np.full_like(array, full_value, [, dtype])

```python
import cv2
img=cv2.imread('image/read_color.jpg')
img
```

```python
a=np.empty_like(img) # 배열의 크기만 같고 내용은 무작위로 들어감
a
```

```python
a=np.zeros_like(img) # 보통 아래와 같이 원하는 초기값을 가지고 생성하는 경우가 많음 
a
```

```python
b=np.zeros_like(img)
b
```

```python
print(img.shape)
print(a.shape)
print(b.shape)
# 차원을 비교하면 모두 같은 크기임
```

### np의 난수 발생 메서드에는 차원을 지정해서 다차원 배열의 난수를 만들 수 있음
- rand():  0~1 사이의 난수를 발생
- randn(): 평균이 0, 분산이 1인 정규 분포(표준 정규 분포)를 따르는 난수를 발생

- 이미지에서 난수는 노이즈를 생성할 때 많이 사용
  - 일정한 분포를 가진 노이즈를 생성할 때 randn()을 사용
- randint(): 정수 난수 발생, 파이썬 메서드의 randint()와 다르게 끝값이 포함되지 않음

```python
d=np.random.rand(2,3) # 0~1 사이값으로 [2,3]
d
```

```python
e=np.random.randn(2,3) # 표준 정규 분포를 따르는 난수
e
```

```python
f=np.random.randint(0,256,(3,3)) # 0~255까지의 난수를 3행 3열로 생성
f
```

### 배열 속성 변경
- ndarray.astype(dtype): dtype을 강제로 바꿈
- 프로그램에 명시적으로 이 타입으로 변경한다 것을 보여주는 겨웅가 아니라면 보통 astype()을 써서 변경함

```python
a=np.arange(5)
a
```

```python
b=a.astype(np.float64)
b
```

```python
c=np.uint8(a)
c
```

```python
a=np.arange(6)
a
```

### 배열의 차원 변경
- ndarray.reshape(newshape)
- numpy.reshape(ndarray, newshape)

단, 변경한 차원의 요소 수 합이 원래 배열의 요소 수와 같아야 함

```python
b=a.reshape(2,3)
b
```

```python
c=np.reshape(a,(2,3))
c
```

```python
d=np.arange(100).reshape(2,-1) # reshape(2,50)과 동일함 -> -1을 사용하면 요소 수 계산을 하지 않아도 됨 
d
```

```python
e=np.arange(100).reshape(-1,5)
e
```

### 1차원 배열로 바꾸기

```python
a=np.zeros((2,3))
a
```

```python
b=a.reshape((6,)) # 요소의 총 개수를 센 후에 입력해야 함
b
```

```python
c=a.reshape(-1) # 요소의 수를 세지 않아도 됨
c
```

```python
d=np.ravel(a) # ravel() 함수는 몇 차원의 배열이든 상관없이 1차원으로 바꿔줌, 또한 요소의 수를 세지 않아도 됨
d
```

### 브로드캐스팅 연산
- 차원의 수가 다른 배열 간의 이항 연산이나 크기가 다른 배열 간의 이항 연산을 가능하게 하는 연산
- 배열과 스칼라값의 연산이 많이 사용됨

### 브로드캐스트 3규칙 요약
- 작은 차원의 배열은 왼쪽을 1로 채워서 차원을 맞춘다.
- 크기가 1인 차원은 늘려서 다른 배열의 크기에 맞춘다.
- 양쪽 모두 크기가 다르고 1도 아니면 불가능하다.

```python
a=np.arange(5)
a
```

```python
a+5
a-2
a**2
```

```python
b=np.arange(6).reshape(2,-1)
b
```

```python
b*2
```

```python
a>2
```

```python
a=np.arange(24).reshape(2,3,4)
b=np.arange(12).reshape(3,4) # (3,4) -> (1,3,4) -> (2,3,4)
print(a)
print(b)
```

```python
a+b
```

```python
c=np.arange(3).reshape(3,1) # (3,1) -> (1,3,1) -> (2,3,4)
c
```

```python
a+c
```

```python
d=np.arange(6).reshape(2,3,1)
d
```

```python
e=np.arange(6).reshape(2,3) # == (1,2,3)
e
```

```python
# d+e -> 차원의 크기가 맞지 않아서 오류가 발생함
```

```python
f=np.arange(6).reshape(3,2)
f
```

```python
d+f
```
