```python
from tkinter import Tk, Button, Label, filedialog
from PIL import Image, ImageTk
from engine import mask_generator, placer, collage_generator

class MainWindow:
    def __init__(self):
        self.root = Tk()
        self.root.title("디지털 콜라주 생성기")
        self.root.geometry("400x200")
        self.preview_label = Label(self.root)
        self.preview_label.pack()

    def open_image(self):
        file_path = filedialog.askopenfilename()
        if not file_path:
            return

        img = image_loader.load_image(file_path)
        img = image_loader.resize_image(img, 300)
        mask = mask_generator.create_mask(img, "circle")
        masked = mask_generator.apply_mask(img, mask)
        piece = collage_generator.random_transform(masked)
        canvas = placer.random_place(piece)

        img_tk = ImageTk.PhotoImage(Image.fromarray(canvas))
        self.preview_label.configure(image=img_tk)
        self.preview_label.image = img_tk

    def run(self):
        Button(self.root, text="이미지 선택", command=self.open_image).pack()
        self.root.mainloop()
```
